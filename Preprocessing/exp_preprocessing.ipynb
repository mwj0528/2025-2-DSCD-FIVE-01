```python
import pandas as pd
import numpy as np
df = pd.read_csv('품목_full.csv', encoding='utf-8-sig')
df.shape
```

```python
def remove_punctuations(text):
    # 1. NaN 또는 비문자열 타입 체크: TypeError 방지
    if pd.isna(text) or not isinstance(text, str):
        return text
    
    result = text.replace('제', '')
    result = result.replace('HSK', '')
    result = result.replace('.', '')
    result = result.replace('-', '')
    
    # 4. 앞뒤 공백 정리
    return result.strip()
```

```python
df['HSCode'] = df['HSCode'].apply(remove_punctuations)
df
```

```python
# 1. 유효성 조건 정의
# a. 문자열 길이가 10인지 확인 (.str.len() == 10)
# b. 문자열이 모두 숫자로만 구성되어 있는지 확인 (.str.isnumeric())

condition_valid_code = (
    df['HSCode'].str.len() == 10
) & (
    df['HSCode'].str.isnumeric()
)

# 2. 유효하지 않은 행의 개수 확인
deleted_rows_count = len(df) - condition_valid_code.sum()
print(f"🚨 유효하지 않은 HS 코드를 포함한 행: {deleted_rows_count}개 삭제 예정.")
```

```python
# 3. 유효한 행만 남기고 삭제
df_cleaned = df[condition_valid_code].copy()

print(f"✅ 최종 데이터프레임 크기: {len(df_cleaned)} 행.\n", 29941-1714)
```

```python
def remove_brackets(text):
    # 1. NaN 또는 비문자열 타입 체크: TypeError 방지
    if pd.isna(text) or not isinstance(text, str):
        return text
    
    result = text.replace('[', '')
    result = result.replace(']', '')
    result = result.replace('\'', '')
    
    # 4. 앞뒤 공백 정리
    return result.strip()

df_cleaned['설명_filtered'] = df_cleaned['설명_filtered'].apply(remove_brackets)
df_cleaned['사유_filtered'] = df_cleaned['사유_filtered'].apply(remove_brackets)
df_cleaned['이름_filtered'] = df_cleaned['이름_filtered'].apply(remove_brackets)

df_cleaned
```

```python
df_cleaned.drop(columns = '번호', inplace = True)
df_cleaned.to_csv('examples_cleaned.csv', index = False)
df_cleaned
```
