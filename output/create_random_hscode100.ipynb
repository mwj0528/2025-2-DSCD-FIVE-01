```python
df = pd.read_excel('/Users/zoohunn/Desktop/4-2/데사캡디/data/관세청_HS부호_20250101 (1).xlsx')
print(df.shape)
df.head()
```

```python
# '한글품목명' 열에서 '기타'가 포함된 행의 개수
count_기타 = df['한글품목명'].astype(str).str.contains('기타', na=False).sum()

print(f"'기타'가 포함된 행의 개수: {count_기타}개")
```

```python
df_filtered = df[~df['한글품목명'].str.contains('기타', na=False)]
df_filtered.shape
```

```python
df_filtered = df_filtered[df_filtered['HS부호'].astype(str).str.len() == 10]
```

```python
sample_df = df_filtered.sample(n=100, random_state=42)
sample_df.head()
```

```python
sample_df.shape
```

```python
sample_df = sample_df.reset_index(drop=True)  
sample_df['번호'] = range(1, len(sample_df) + 1)

cols = ['번호'] + [c for c in sample_df.columns if c != '번호']
sample_df = sample_df[cols]


sample_df.head()
```

```python
cols_to_keep = [
    '번호',
    'HS부호', 
    '적용시작일자', 
    '적용종료일자', 
    '한글품목명', 
    '영문품목명', 
    '한국표준무역분류명', 
    '성질통합분류코드명'
]

df_selected = sample_df[cols_to_keep].copy()

df_selected['상품명'] = ''
df_selected['상품설명'] = ''

cols_order = cols_to_keep + ['상품명', '상품설명']
df_selected = df_selected[cols_order]

df_selected.head()
```

```python
df_selected.to_excel("/Users/zoohunn/Desktop/4-2/데사캡디/data/HScode_랜덤100개_기타제외.xlsx", index=False)
```
