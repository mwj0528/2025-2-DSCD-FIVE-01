```python
import pandas as pd
```

```python
df = pd.read_excel('/Users/zoohunn/Desktop/HScode_100개_부모코드4:6행빼낸_data.xlsx')
df.head()
```

```python
df.tail()
```

```python
# 10자리 코드 분리 (level이 NaN인 것)
df_10digit = df[df['level'].isna()].copy()
df_10digit = df_10digit.drop(columns=['category', 'code', 'description', 'hs10', 'level','parent_code','parent_of','번호','원본번호'])
df_10digit
```

```python
# 부모 코드 분리 (level이 4.0 또는 6.0인 것)
df_parent = df[df['level'].notna()].copy()
df_parent = df_parent.drop(columns=['HS10','HS부호','hs10','번호','성질통합분류코드명','영문품목명', '원본번호', '한글품목명'])
df_parent
```

```python
# 정수형 변환 (매칭을 위해)
df_10digit['HS10_int'] = df_10digit['HS10'].astype('int64')
df_parent['parent_of_int'] = df_parent['parent_of'].fillna(0).astype('int64')

# 사용자_상품설명 열 생성
df_10digit['사용자_상품설명'] = ''
df_10digit
```

```python
# 각 10자리 코드에 대해 부모 정보 추가
for idx, row in df_10digit.iterrows():
    hs10_code = row['HS10_int']
    성질통합분류코드명 = row['성질통합분류코드명'] if pd.notna(row['성질통합분류코드명']) else ''
    
    # 매칭되는 부모 레코드 찾기
    matched_parents = df_parent[df_parent['parent_of_int'] == hs10_code]
    
    # Level별로 분리
    level_4_desc = ''
    level_6_desc = ''
    
    for _, parent_row in matched_parents.iterrows():
        level = parent_row['level']
        description = parent_row['description'] if pd.notna(parent_row['description']) else ''
        
        if level == 4.0:
            level_4_desc = description
        elif level == 6.0:
            level_6_desc = description
    
    # 새로운 형식으로 조합
    parts = []
    
    if 성질통합분류코드명:
        parts.append(성질통합분류코드명)
    
    if level_4_desc:
        parts.append(f"대분류: {level_4_desc}")
    
    if level_6_desc:
        parts.append(f"중분류: {level_6_desc}")
    
    # ". "로 연결
    사용자_상품설명 = ". ".join(parts) + "." if parts else ''
    
    df_10digit.at[idx, '사용자_상품설명'] = 사용자_상품설명
```

```python
df_10digit['사용자_상품명'] = df_10digit['한글품목명']

# 열 순서 변경 (사용자_상품명을 사용자_상품설명 앞으로)
cols = df_10digit.columns.tolist()
cols.remove('사용자_상품명')  
idx = cols.index('사용자_상품설명')  
cols.insert(idx, '사용자_상품명')  
df_10digit = df_10digit[cols]  
df_10digit
```
